```python
%matplotlib inline
import matplotlib.pyplot as plt
from __future__ import print_function
from getdist import plots, MCSamples
import getdist
import numpy as np
```

```python
samples = np.loadtxt("../../dat/nbar_gmf_Mr21.mcmc_chain.dat")[250*2500:]
print(samples.shape)
```

```python
import corner
plot_range = np.zeros((5,2))

plot_range[:,0] = [10.5, np.log(0.2), 11.8, 0.85, 13.]
plot_range[:,1] = [13.5, np.log(0.8), 13.8, 1.45, 15.5]

#samples[:,2] = np.exp(samples[:,2])
#ndim, nsamples = 5, 10000
#samples = np.random.randn(ndim * nsamples).reshape([nsamples, ndim])
fig = corner.corner(
            samples,
            truth_color='#ee6a50',
            labels=[
                r'$\mathrm{\logM_{0}}$',
                r'$\mathrm{\log\sigma_{\logM}}$',
                r'$\mathrm{\logM_{min}}$',
                r'$\mathrm{\alpha}$',
                r'$\mathrm{\logM_{1}}$'
                ],
            label_kwargs={'fontsize': 15},
            range=plot_range,
            quantiles=[0.16,0.5,0.84],
            show_titles=True,
            title_args={"fontsize": 12},
            plot_datapoints=False,
            fill_contours=True,
            levels=[.68,.95],
            bins=16,
            smooth=1.)

#figure.savefig("corner.png")
```
